```r
# Load precision scores MLP
precision_file <- file.path("..", "inputs/cpg0004_profiles", "MLP_mAP_10.csv")

precision_cols <- readr::cols(
  compound = readr::col_character(),
  AP = readr::col_double()
)

# Load and process data for plotting
precision_mlp_df <- readr::read_csv(precision_file, col_types = precision_cols) 

# Load precision scores average
precision_file <- file.path("..", "inputs/cpg0004_profiles", "average_mAP_10.csv")

# Load and process data for plotting
precision_av_df <- readr::read_csv(precision_file, col_types = precision_cols) 

# Add label 
precision_av_df <- precision_av_df %>%
    mutate(average = precision_av_df$AP)


# Add label 
precision_mlp_df <- precision_mlp_df %>%
    mutate(model = precision_mlp_df$AP)

dose10_df <- precision_mlp_df %>%
    mutate(average = precision_av_df$AP)

dose10_df <- dose10_df %>%
    mutate(dose = '10 uM')
```

```r
# Load precision scores MLP
precision_file <- file.path("..", "inputs/cpg0004_profiles", "MLP_mAP_3.csv")

precision_cols <- readr::cols(
  compound = readr::col_character(),
  AP = readr::col_double()
)

# Load and process data for plotting
precision_mlp_df <- readr::read_csv(precision_file, col_types = precision_cols) 

# Load precision scores average
precision_file <- file.path("..", "inputs/cpg0004_profiles", "average_mAP_3.csv")

# Load and process data for plotting
precision_av_df <- readr::read_csv(precision_file, col_types = precision_cols) 

# Add label 
precision_av_df <- precision_av_df %>%
    mutate(average = precision_av_df$AP)


# Add label 
precision_mlp_df <- precision_mlp_df %>%
    mutate(model = precision_mlp_df$AP)

dose3_df <- precision_mlp_df %>%
    mutate(average = precision_av_df$AP)

dose3_df <- dose3_df %>%
    mutate(dose = '3.33 uM')
```

```r
library(dplyr)

long_annotation <- "dehydrogenase inhibitor/inositol monophosphatase inhibitor"
new_annotation <- "dehydrogenase inhibitor"

panel_d_df <- panel_d_df %>%
  mutate(compound = ifelse(compound == long_annotation, new_annotation, compound))
```

```r
library(ggplot2)
library(ggrepel)

color_logic <- (
    panel_d_df$average > 0.4 |
    panel_d_df$model > 0.4
    )


panel_d_gg <- (
    ggplot(panel_d_df, aes(x = model, y = average, size = count, color = dose))
    + theme_bw()
    + geom_abline(intercept = 0, slope = 1, linetype = "dashed", color = "black")
    + theme(text = element_text(size = 30),
          axis.title.x = element_text(colour = "#701070", size=34),
          axis.title.y = element_text(colour = "#df20df", size=34))
    + xlab("mAP CytoSummaryNet (-)")
    + ylab("mAP average profiling (-)")
    + labs(size="Number of compounds\nin MoA class")
    + geom_text_repel(
        data = subset(panel_d_df, color_logic),
        arrow = arrow(length = unit(0.020, "npc")),
        segment.size = 0.7,
        segment.alpha = 0.6,
        size = 8,
        fontface = "italic",
        box.padding = 0.9,
        point.padding = 0.8,
        aes(
            x = model,
            y = average,
            label = compound,
        )
    )
    + geom_point(alpha = 0.8)
    #+ theme_gray(base_size = 34)
#     + scale_color_manual(name = "Dose", values = dose_colors)
    + scale_x_continuous(breaks = seq(0, 1, 0.2), limits = c(0, 1.2))
    + scale_y_continuous(breaks = seq(0, 1, 0.2), limits = c(0, 1.2))
    #+ scale_size(breaks = seq(10, 80, 30), limits=c(7, 90))
    + scale_size(range = c(0, 10), limits=c(1, 90))
    + guides(
        color = guide_legend(override.aes = list(size = 15, alpha = 1)),
        text = guide_legend(override.aes = list(alpha = 0))
    )
)

panel_d_gg
```
